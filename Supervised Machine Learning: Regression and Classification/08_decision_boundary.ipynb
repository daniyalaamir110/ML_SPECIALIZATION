```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDClassifier
```

```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1]).reshape(-1,1)
```

```python
positive_points = X[y.flatten() == 1]
negative_points = X[y.flatten() == 0]
plt.scatter(positive_points[:, 0], positive_points[:, 1], c="b", s=32, marker="x", label="y = 1")
plt.scatter(negative_points[:, 0], negative_points[:, 1], c="r", s=32, marker="x", label="y = 0")
plt.legend()
plt.xlabel("x0")
plt.ylabel("x1")
plt.show()
```

```python
model = SGDClassifier()
model.fit(X, y.flatten())
w = model.coef_[0]
b = model.intercept_[0]
```

```python
min_x0 = np.min(X[:, 0])
max_x0 = np.max(X[:, 0])
db_x = np.linspace(min_x0, max_x0, 10)
db_y = (-b - w[0] * db_x) / w[1]
positive_points = X[y.flatten() == 1]
negative_points = X[y.flatten() == 0]
plt.scatter(positive_points[:, 0], positive_points[:, 1], c="b", s=32, marker="x", label="y = 1")
plt.plot(db_x, db_y, c="g", label="Decision boundary")
plt.scatter(negative_points[:, 0], negative_points[:, 1], c="r", s=32, marker="x", label="y = 0")
plt.legend()
plt.xlabel("x0")
plt.ylabel("x1")
plt.show()
```
