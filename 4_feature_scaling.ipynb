```python
def scatter_all(x, y, features, output):
  n = len(features)
  fig, ax = plt.subplots(1, n, figsize=(3 * n, 3), sharey=True)
  for i in range(len(ax)):
      ax[i].scatter(x[:, i], y)
      ax[i].set_xlabel(features[i])
  ax[0].set_ylabel(output)
  plt.show()


def scatter_comp(x_old, x_new, y_old, y_new, xlabel, ylabel):
  fig, ax = plt.subplots(1, 2, figsize=(6, 3))
  ax[0].scatter(x_old, y_old)
  ax[0].set_xlabel(xlabel)
  ax[0].set_ylabel(ylabel)
  ax[1].scatter(x_new, y_new)
  ax[1].set_xlabel(xlabel)
  plt.show()
```

```python
scatter_all(X_train, y_train, X_features, y_output)
```

```python
def zscore_normalization(x):
  mean = np.mean(x, axis=0)
  std = np.std(x, axis=0)

  X_norm = (x - mean) / std

  return X_norm


def standardization(x):
  maxs = np.max(x, axis=0)

  X_standard = x / maxs

  return X_standard


def mean_normalization(x):
  mean = np.mean(x, axis=0)
  mins = np.min(x, axis=0)
  maxs = np.max(x, axis=0)

  X_norm = (x - mean) / (maxs - mins)

  return X_norm
```

```python
X_znorm = zscore_normalization(X_train)
scatter_all(X_znorm, y_train, X_features, y_output)
```
